```python
import scipy.stats as stats
import pylab

arr = df.to_numpy()
print(len(arr))
l= len(arr)
k=[]
for i in range(l):
    kk=arr[i][10] 
    k.append(kk)
#print(k)
#measurements = np.random.normal(loc = 20, scale = 5, size=100)   
k=np.array(k)
stats.probplot(k, dist="norm", plot=pylab)
pylab.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
corr = df.corr(method='pearson')
print(corr)
```

```python
ax = sns.heatmap(
    corr, 
    vmin=-1, vmax=1, center=0,
    cmap=sns.diverging_palette(100, 440, n=400),
    square=True
)
ax.set_xticklabels(
    ax.get_xticklabels(),
    rotation=45,
    horizontalalignment='right'
);
```
